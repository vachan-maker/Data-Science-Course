```python
#from sklearn import iris
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris_df = pd.read_csv('/content/Iris.csv')
iris_df.head()
```

1.Training the model - we need training data
2. Testing the model - we need to test the data

Split our data into: Training data and testing data

```python
from sklearn.model_selection import train_test_split
#to split data into training data and testing data
```

We know our data has both input and output

Input(X): Featuure - relevant columns which can answer our questions are called features

Output(Y): Target - Only one column is output column


```python
#identify input columns from the dataset
iris_df.head(2)
#features are sepal length, sepal width, petal length, petal width
X= iris_df.iloc[:,1:5] #all rows and 1to 5 columns
X.head()
```

```python
#identify the output column or target column from the dataset
#Target: [species]
Y = iris_df.iloc[:,5] #all rows in the last column
Y.head()
```

```python
#X feature set
#Y output or target
#X test features for testing
#Y_test target for testing feature
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42) #random state. To ask the data in between for training
```

```python
#standardization for features - to make everything in same range
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train) #fit everything in the range of 0 to 1
X_test = scaler.transform(X_test)
```

```python
#apply ML algorithm
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
