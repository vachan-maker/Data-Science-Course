# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('/content/car_details.csv')
```

```python
cars.head(3)
#top 3 rows
```

```python
cars.columns
#differnet columns
```

```python
#information of all the columns
cars.info()
```

```python
#statistical details about numerical columns
cars.describe()
```

```python
#statistical details of object type columns
cars.describe(include = 'object')
```

```python
cars.Make.value_counts() #find out the count of different makes of cars
```

```python
cars['Vehicle Style'].value_counts()
```

```python
cars['Transmission Type'].value_counts()
```

```python
plt.figure(figsize=(20,20))
cars['Engine Fuel Type'].value_counts().plot(kind='pie')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y = 'Engine Fuel Type', x = 'MSRP', data = cars, palette = 'Set2')
plt.show()
```

```python
#to create a box plot between Engine HP and Engine Cylinders
plt.figure(figsize = (12,10))
sns.boxplot(y='Engine HP', x = 'Engine Cylinders', data = cars, palette = 'viridis')
plt.show()
```

```python
#how many cars have engine cylinders = 0
cars[cars['Engine Cylinders']==0]['Engine Fuel Type'].value_counts()
```

```python
cars[cars['Engine Cylinders']==16]['Make'].value_counts()
```

```python
#value counts of driven wheels
sns.countplot(x = 'Driven_Wheels', data =cars, palette = 'viridis')
```

```python
sns.countplot(y='Year', data = cars, palette = 'viridis')
```

```python
cars['Vehicle Size'].value_counts()
```

```python
sns.countplot(y='Vehicle Style', data=cars)
```

```python
#Fuel Consumption
#histo
plt.figure(figsize = (12,20))
sns.histplot(data=cars, x = 'highway MPG', hue = 'Engine Cylinders',palette = 'viridis')
plt.xlim([0,100])
```

```python
sns.scatterplot(data=cars, y = 'city mpg', x = 'highway MPG', hue = 'Engine Cylinders', palette = 'inferno')
plt.xlim([0,60])
plt.ylim([0,60])
```

```python
#see only EVs
sns.scatterplot(data=cars, y = 'city mpg', x = 'highway MPG', hue = 'Engine Cylinders', palette = 'inferno')
plt.xlim(60,150)
```

```python
brand = cars.groupby(by='Make')['MSRP'].mean().sort_values(ascending = False)
brand.head()

#convert this to a dataframe
brand_df = pd.DataFrame(brand)
plt.figure(figsize=(15,15))
sns.barplot(data = brand_df, x= 'Make', y = 'MSRP')
plt.xticks(rotation=45)
plt.show()
```
