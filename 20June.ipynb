```python
#rowwise manipulation - loc function by name
#iloc - rocates rows by the numeric index
import pandas as pd
df2 = pd.read_csv('IMDB-Movie-Data.csv')
df2.set_index('Title', inplace=True) #setting title as the inedex
df2.loc['Guardians of the Galaxy']
```

```python
df2.iloc[1,4]
```

```python
#locate the row with details of movie arrival
df2.loc['Arrival']
```

```python
#locate the row details of row no 440 to 445
df2.iloc[439:446]
```

```python
#display details of seperate rows at once
df2.iloc[[2,3,4,5,6]]
```

```python
#Print info of movies with rating greater than 8.5
df2[df2['Rating']>8.5]
```

```python
#multiple conditions & symbol or | symbol
df2[df2['Director'] == 'James Gunn']
```

```python
#list movies of james gunn with less than 7 as rating
df2[(df2['Director'] == 'James Gunn') & (df2['Rating'] >7)]
```

```python
#list movies between years 2010 to 2013
#movies between years 2011 to 2015 with rating greater >8
df2[(df2['Year'] > 2010) & (df2['Year']<2013)]
```

```python
df2[(df2['Year'] >= 2011) & (df2['Year']<+ 2015) & (df2['Rating'] > 8)]
```

```python
df2[(df2['Year'] >= 2008) & (df2['Year']<= 2014) & (df2['Rating'] > 7) &(df2['Revenue (Millions)'] > 500)]
```

```python
#calculate the average rating
avg_rating = df2['Rating'].mean()
avg_rating
```

```python
#show max rating
max_rating = df2['Rating'].max()
max_rating
```

```python
df2[df2['Rating'] == 9.0]
#to find the movie with rating 9
```

```python
#group movies by directors and also calculate avg rating each director
df2.groupby('Director')['Rating'].mean()
```

```python
#display the top 5 directors with highest avg rating
df2.groupby('Director')['Rating'].mean().sort_values(ascending = False).head(5)
```

```python
df2.groupby('Genre')['Revenue (Millions)'].mean().sort_values(ascending = False).head(10)
```
