## Seaborn Visualization

#### Scatterplot:
we will be using lmplot() function to scatter plot

(data = data frame name, x = column1, y = column2)

```python
sn.lmplot(x = 'Attack', y = 'Defense', data=df, fit_reg = False, hue = 'Stage') #to remove regression line
#hue should be used mostly with categorical columns
```

## Violin plot
useful alternative of boxplot
- shows distribution through thickness

```python
sn.set_style('whitegrid')
plt.figure(figsize = (12,10))
sn.violinplot(x = 'Type 1', y = 'Attack', data = df, palette = 'pastel')
```

```python
df2 = pd.read_csv('/content/IMDB-Movie-Data.csv', index_col = 0, encoding = 'latin')
sn.set_theme(style = 'darkgrid')
sn.set_palette('Set2')
plt.title('Revenue vs Rating')
# x =rating
# y = revenue
sn.scatterplot(x = 'Rating', y = 'Revenue (Millions)', data=df2, hue = 'Year')
```

```python
plt.figure(figsize=(12,30))
plt.ylabel('Genre')
plt.xlabel('Rating')
sn.boxplot(x ='Rating' , y ='Genre' , data =df2, palette = 'Set2', orient = 'h')
```

```python
#countplot
plt.title("Movies per Year")
sn.countplot(y = "Year", data = df2,palette = 'viridis')
```

```python
#bar plot
#top 10 directors with max revenue

#Step1 group by directors
# Step 2 - Sort the directors by revenue and select the top 10
#step 3: use barplot()

#dataframe or data series
director_revenue = [[df.groupby('Director')['Revenue (Millions)'].sum().sort_values(ascending = False).head(10)]]
director_revenue
```
