```python
import pandas as pd
import matplotlib.pyplot as plt
```

Dataset:  

* **#**: ID for each pokemon
* **Name**: Name of each pokemon
* **Type 1**: Each pokemon has a type, this determines weakness/resistance to attacks
* **Type 2**: Some pokemon are dual type and have 2
* **Total**: sum of all stats that come after this, a general guide to how strong a pokemon is
* **HP**: hit points, or health, defines how much damage a pokemon can withstand before fainting
* **Attack**: the base modifier for normal attacks (eg. Scratch, Punch)
* **Defense**: the base damage resistance against normal attacks
* **SP Atk**: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
* **SP Def**: the base damage resistance against special attacks
* **Speed**: determines which pokemon attacks first each round

```python
df = pd.read_csv('/content/Pokemon.csv', index_col=0,encoding='latin')
df.head()
```

## **Seaborn Plotting Functions**

### Scatterplot:

```python
# sns.lmplot()
# (data = df_name, x = column1, y= column2)

sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # no regression line
           hue = 'Stage') # color by evolution stage

# hue argument is useful bcs it allows you to express a third column of information using color
```

```python
sns.scatterplot(x='Attack', y='Defense', data=df,
                 hue = 'Stage') # color by evolution stage
```

```python
#boxplot

plt.figure(figsize=(12,10))
sns.boxplot(data=df)
```

## Change theme of seaborn graphs:


there are 5 preset themes:
1. darkgrid : grey bckrgnd + gridlines
2. whitegrid : White bckgrnd + gridlines
3. dark : grey bckgrnd
4. white : White bckgrnd
5. ticks: when need for extra structure to plots

```python
sns.set_style('dark')
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False)
```

## Violin plot

* useful alternative of boxplot
* shows distribution (through thickness of the violin) instead of summary statistics


```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,10))
sns.violinplot(x='Type 1', y='Attack', data=df, palette='husl')
```

## Seabron on IMDB dataset

```python
movies_df = pd.read_csv('/content/IMDB-Movie-Data.csv', index_col='Title')
```

```python
#reading the first 2 rows of the dataframe
movies_df.head(2)
```

```python
#setting a theme

sns.set_theme(style='darkgrid')  # 'white' ,'whitegrid','dark', 'darkgrid' , 'ticks'
sns.set_palette('pastel')
# plot 1: Histogram plot : one column plot as x-axis , y-axis : frequency count
plt.figure(figsize=(12,10))
sns.histplot(data=movies_df, x='Rating', kde=True,bins=10)
plt.title('IMDB Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('No. of movies')
plt.show()
```

```python
movies_df.columns
```

```python
movies_df = pd.read_csv('/content/IMDB-Movie-Data.csv')
movies_df.head(2)
```

```python
# Scatter plot : sns.scatterplot(x= 'col1', y= 'col2', data = df_name)
# x = rating   y= revenue
# show title, x-label, y-label, in whitegrid theme

plt.figure(figsize=(12,10))
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=movies_df,hue='Year')
sns.set_style('whitegrid')
sns.set_palette('pastel')
plt.title('Revenue vs Rating')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.show()
```

```python
# Boxplot for Ratign distribution by genre
# sns.boxplot(x= 'col1',y='col2',data=df)

plt.figure(figsize=(12,30))
sns.boxplot(y='Genre', x='Rating', data=movies_df,palette='Set2', orient='h')
plt.ylabel('Genre')
plt.title('Rating Distbn')
```

```python
movies_df.columns
```

```python
# countplot: show number of movies released each year

sns.countplot(y='Year', data=movies_df,palette='viridis')
plt.title('Movies per year')
```

```python
# BAR plot: top 10 directors with max revenue

# Step 1: group data by directors (groupby)
# Step 2:Sort the directors by total revenue and select the top 10 .
# Step 3: use barplot()

director_revenue=movies_df.groupby('Director')['Revenue (Millions)'].sum().sort_values(ascending=False).head(10)

## convert this to dataframe.
## col1 : Director   col2: Revenue
## sns.barplot()

director_rev=pd.DataFrame(director_revenue)
director_rev.reset_index(inplace=True)  #

plt.figure(figsize=(12,10))
sns.barplot(x='Director',y='Revenue (Millions)', data=director_rev)
# xti
```

## **Dataset 3: TIPS dataset**

```python
tips = sns.load_dataset('tips')  #import TIPS dataset into file
tips.head()
```

```python
#1. countplot : number of females and males customers in the restaturant
plt.figure(figsize=(8,6))
sns.countplot(x='sex', data=tips,palette='pastel')
plt.title('Gender')
plt.show()
```

```python
#2. countplot: number of smokers
plt.figure(figsize=(8,6))
sns.countplot(x='smoker', data=tips,palette='pastel')
plt.title('Count of smokers')
plt.show()
```

```python
#3. scatterplot of totalbill and tip. hue=smokers
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
plt.title('Total bill vs tip')
```

```python
#4. scatterplot of totalbill and tip . hue = sex
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Total bill vs tip')
```

```python
#5. barplot : x=day y= tips  hue = sex
sns.barplot(x='day', y='tip', data=tips, hue='sex',palette='viridis')
```

```python
#6. histplot of tips
sns.histplot(x='tip', data=tips,bins=5,kde=True,color='magenta')
plt.title('Histogram of tips')
```
