Typically we have two options
1. Delete rows that contain missing data
2. Replace them with a value which is called imputation
Deleting that many rows might affect the data so we are not going to delete them instead we are going to replace them with a value

There are several ways to impute
1. One of the most simplest ways is by replacing null values in each column with the column name or mode.

2. For credit score, we will find the mean credit score for a particulat category ie upper class, middle class and put the mean value in place of the null values. **Credit score** is heavily affected by income.

We can do a group by function to see how the mean values for each income group

```python
df.groupby('income')['credit_score'].mean() #finding the mean credit score by income
```

To impute the missing values for credit score with mean credit score for each income group, the simplest way to do this is by creating a function so we don't have to repeate the code for each individiual

```python
def impute_credit_score(income_classes):
    for income_class in income_classes:
        #create a subset of the dataframe to use a as a filter
        #calculate the mean for the income group
        #fill the missisng values with mean values
        mask = df['income'] == income_class
        mean_value = df[df['income']==income_class]['credit_score'].mean()
        df.loc[mask, 'credit_score'] = df.loc[mask,'credit_score'].fillna(mean_value)
        
```

```python
income_groups=['poverty','working class', 'middle class', 'upper class']
impute_credit_score(income_groups)
```

```python
df.isnull().sum()
```

Lets tackle **annual_mileage**

```python
#Check for mean annual milage for different driving experience
df.groupby('driving_experience')['annual_mileage'].mean()
```

The mean of different groups in the driving experience do not vary widely too much. So we can simply impute the null values using the column mean

#calculate the mean annual mileage column


```python
mean_mileage = df['annual_mileage'].mean()
mean_mileage
```

```python
df['annual_mileage'] = df['annual_mileage'].fillna(mean_mileage)
```
