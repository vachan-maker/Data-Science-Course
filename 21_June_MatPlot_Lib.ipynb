```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6869]
steps_lastweek = [9788,8710,5303,17680,21390,4002,5223]

plt.plot(days, steps_walked, 'x-r') #plot two things
plt.plot(days, steps_lastweek, 'v-y')

plt.title('This week vs Last Week')
plt.xlabel('Days')
plt.ylabel('Steps_walked')

plt.grid(True)

plt.legend(["This-week", "Last-week"])
plt.show()
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6869]
steps_lastweek = [9788,8710,5303,17680,21390,4002,5223]

fig,ax = plt.subplots()
#figure, axes(ax)
ax.plot(days, steps_walked, 'x-r') #plot two things
ax.plot(days, steps_lastweek, 'v-y')

ax.set_title('This week vs Last Week')
ax.set_xlabel('Days')
ax.set_ylabel('Steps_walked')

ax.grid(True)

ax.legend(["This-week", "Last-week"])
plt.show()
```

# Create Subplots


```python

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6869]
steps_lastweek = [9788,8710,5303,17680,21390,4002,5223]

fig, axs = plt.subplots(2,2)  # Creates two plots
axs[0,0].plot(days, steps_walked, 'x-r') #plot two things
axs[0,0].plot(days, steps_lastweek, 'v-y')

axs[0,0].set_title('This week vs Last Week')
axs[0,0].set_xlabel('Days')
axs[0,0].set_ylabel('Steps_walked')

axs[0,0].grid(True)

axs[0,0].legend(["This-week", "Last-week"])

axs[0,1].bar(days, steps_walked, 'x-r') #plot two things
axs[0,1].bar(days, steps_lastweek, 'v-y')

axs[0,1].set_title('This week vs Last Week')
axs[0,1].set_xlabel('Days')
axs[0,1].set_ylabel('Steps_walked')

axs[0,1].grid(True)

axs[0,1].legend(["This-week", "Last-week"])
plt.show()
# One row and one column
```

```python
students = ['Angel', 'Barabara', 'Colin', 'David', 'Elga', 'Fiona']
test1 = [78, 54, 79, 92, 83.8, 94.5]
test2= [84.4, 67.3, 71.2,91.6,89.2,12]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(students, test1, 'x-r')
axs[0, 0].plot(students, test2, 'v-y')
axs[0, 0].set_title('This week vs Last Week (Line Plot)')
axs[0, 0].set_xlabel('Students')
axs[0, 0].set_ylabel('Scores')
axs[0, 0].grid(True)
axs[0, 0].legend(["This-week", "Last-week"])


axs[0, 1].bar(students, test1)
axs[0, 1].bar(students, test2)
axs[0, 1].set_title('This week vs Last Week (Line Plot)')
axs[0, 1].set_xlabel('Students')
axs[0, 1].set_ylabel('Scores')
axs[0, 1].grid(True)
axs[0, 1].legend(["This-week", "Last-week"])

axs[1, 0].stem(students, test1)
axs[1, 0].stem(students, test2)
axs[1, 0].set_title('This week vs Last Week (Line Plot)')
axs[1, 0].set_xlabel('Students')
axs[1, 0].set_ylabel('Scores')
axs[1, 0].grid(True)
axs[1, 0].legend(["This-week", "Last-week"])

axs[1, 1].scatter(students, test1)
axs[1, 1].scatter(students, test2)
axs[1, 1].set_title('This week vs Last Week (Line Plot)')
axs[1, 1].set_xlabel('Students')
axs[1, 1].set_ylabel('Scores')
axs[1, 1].grid(True)
axs[1, 1].legend(["This-week", "Last-week"])

fig.suptitle('Marks Comparison')
fig.legend(['Test1', 'Test2'])
plt.show()
```

You can create a figure object with more than one axes object
